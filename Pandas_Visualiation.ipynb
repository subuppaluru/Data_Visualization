```python
#importing packages
import numpy as np
import pandas as pd

#creating a DataFrame
df = pd.DataFrame(np.random.rand(10, 4), 
                  columns=('col_1', 'col_2', 'col_3', 'col_4'))

#displaying the DataFrame
df
```

#Line plot

```python
df.plot()
```

```python
df.plot(x="col_1", y="col_2")
```

```python
df.plot(subplots=True, figsize=(8, 8))
```

#Bar plot:

```python
df.plot(kind="bar")
```

```python
df.plot.bar(stacked=True)
```

```python
df.plot.barh(stacked=True)
```

#Histogram plot

```python
df.plot.hist()
```

```python
df.plot.hist(stacked=True, bins=20)
```

```python
df.plot.hist(orientation="horizontal", cumulative=True)
```

```python
df.diff().hist()
```

#Box plot

```python
df.plot.box()
```

```python
df.boxplot()
```

```python
df.plot.box(vert=False, positions=[1, 2, 3, 4])
```

#Area plot

```python
df.plot.area()
```

```python
df.plot.area(stacked=False)
```

```python
df.plot.area(stacked=True)
```

#Scatter plot

```python
df.plot.scatter(x='col_1', y='col_3')
```

```python
ax=df.plot.scatter(x="col_1", y="col_3", color="red", marker="*", s=100)
df.plot.scatter(x="col_2", y="col_4", color="orange", s=100, ax=ax)
```

```python
ax = df.plot.scatter(x="col_2", y="col_4", color="orange", s=100, ax=ax)
```

```python
df.plot.scatter(x="col_2", y="col_4", c='col_1', s=100)
```
