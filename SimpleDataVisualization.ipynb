```python
#https://www.kaggle.com/vanshjatana/a-simple-tutorial-to-data-visualization/notebook
import pandas as pd
import matplotlib.pyplot as plt
cp = pd.read_csv("/content/drive/MyDrive/AI/DataVisualization/DV/Placement_Data_Full_Class.csv")
price = pd.read_csv("/content/drive/MyDrive/AI/DataVisualization/DV/house_train.csv")
age = pd.read_csv("/content/drive/MyDrive/AI/DataVisualization/DV/titantic_train.csv")
sales = pd.read_csv("/content/drive/MyDrive/AI/DataVisualization/DV/items.csv")
word = pd.read_csv("/content/drive/MyDrive/AI/DataVisualization/DV/nlp_train.csv")
cases = pd.read_csv("/content/drive/MyDrive/AI/DataVisualization/DV/covid_19_data.csv")
iris = pd.read_csv("/content/drive/MyDrive/AI/DataVisualization/DV/Iris.csv")
cp = cp.fillna(0)
stocks = pd.read_csv("/content/drive/MyDrive/AI/DataVisualization/DV/TSLA.csv")
```

#Bar Plot

##Bar

```python
import plotly.express as px
grs = cp.groupby(["gender"])[["salary"]].mean().reset_index()
fig = px.bar(grs[['gender', 'salary']].sort_values('salary', ascending=False), 
             y="salary", x="gender", color='gender', 
             log_y=True, template='ggplot2')
fig.show()
```

##Horizontal Bar

```python
grs = cp.groupby(["gender"])[["salary"]].mean().reset_index()
fig = px.bar(grs[['gender', 'salary']].sort_values('salary', ascending=False), 
             y="gender", x="salary", color='gender', orientation = 'h')
fig.show()
```

##Stacked Bar

```python
grgs = cp.groupby(["gender","specialisation"])[["salary"]].mean().reset_index()
fig = px.bar(grgs, x="gender", y="salary", color='specialisation', barmode='stack',
             height=400)
fig.show()
```
